# EDA for Manchester City FC datasets

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib as mp
```

```python
# Import the Manchester City dataset for 2011-2012 EPL season
df_2011_2012_season = pd.read_excel('mcfc_data/Premier League 2011-12 Match by Match.xls', sheetname='Players')
```

```python
# List all the column names
#list(df_2011_2012_season.columns)
```

```python
# Some season totals
df_Player_Season_Totals = df_2011_2012_season.groupby('Player Surname').sum()
df_Team_Season_Totals = df_2011_2012_season.groupby('Team').sum()
```

```python
plot_Player_TouB_G = df_Player_Season_Totals.plot(x='Touches open play opp box',y='Goals Open Play',marker='.',linestyle='None')
title('By Player')
xlabel('Touches in Opponent Box')
ylabel('Goals From Open Play')
```

```python
#Linear reg.
x = sm.add_constant(df_Team_Season_Totals['Total Successful Passes All'],prepend=True)
model1 = sm.OLS(df_Team_Season_Totals['Goals Open Play'],x).fit()
intercept, slope = model1.params
```

```python
plot_x = [min(df_Team_Season_Totals['Total Successful Passes All']),max(df_Team_Season_Totals['Total Successful Passes All'])]
y = intercept + slope*plot_x
```

```python
# Plot Open Play Goals v. Total Successful Passes + Linear Reg.
from statsmodels.graphics.api import abline_plot

figure1 = abline_plot(model_results=model1)
ax = figure1.axes[0]
ax.scatter(df_Team_Season_Totals['Total Successful Passes All'], df_Team_Season_Totals['Goals Open Play'])
ax.margins(.1)
#plot_Team_Tou_G = df_Team_Season_Totals.plot(x='Total Successful Passes All',y='Goals Open Play',marker='.',linestyle='None')

ax.annotate(['R^2=',model1.rsquared],xy=(10,-20), xycoords='axes points')

title('By Team')
ylabel('Goals From Open Play')
for i, txt in enumerate(df_Team_Season_Totals.index):
    ax.annotate(txt, (df_Team_Season_Totals['Total Successful Passes All'].iloc[i],df_Team_Season_Totals['Goals Open Play'].iloc[i]), size=10, xytext=(0,0), ha='right', textcoords='offset points')
    
    
```

```python
# Return a player dataset

def player_data(player_info):

    if type(player_info) == type(''):
        df_player = df_2011_2012_season[df_2011_2012_season['Player Surname'] == player_info]
    else:
        df_player = df_2011_2012_season[df_2011_2012_season['Player ID'] == player_info]
    
    df_player.index = pd.to_datetime(df_player['Date'])
    df_player = df_player.sort('Date')
    
    # General Indicators
    df_player['Pass Success Rate'] = df_player['Total Successful Passes Excl Crosses Corners'].cumsum()/(df_player['Total Successful Passes Excl Crosses Corners'].cumsum()+df_player['Total Unsuccessful Passes Excl Crosses Corners'].cumsum())
    df_player['Fouls Won p90'] = df_player['Total Fouls Won'].cumsum()*90/df_player['Time Played'].cumsum()
    
    
    # Forward Indicators
    df_player['GOPtoSOT'] = df_player['Goals Open Play'].cumsum() / df_player['Shots On Target inc goals'].cumsum()
    df_player['SOTtoTouF3rd'] = df_player['Shots On Target inc goals'].cumsum() / df_player['Touches open play final third'].cumsum()
    df_player['SOTtoTou'] = df_player['Shots On Target inc goals'].cumsum() / df_player['Touches'].cumsum()
    
    # Holding Midfielder Indicators

    
    
    return df_player
```

```python
df_Dzeko = player_data('Dzeko')
df_Suarez = player_data(39336)
df_RVP = player_data('van Persie')
df_Gerrard = player_data('Gerrard')
df_Lampard = player_data('Lampard')
df_Balotelli = player_data('Balotelli')
df_Rooney = player_data('Rooney')
df_Aguero = player_data(37572)
```

```python
df_Suarez['Goals'].cumsum().plot(marker='.')
df_Dzeko['Goals'].cumsum().plot(marker='.')
df_RVP['Goals'].cumsum().plot(marker='.')
df_Aguero['Goals'].cumsum().plot(marker='.')
df_Rooney['Goals'].cumsum().plot(marker='.')
df_Balotelli['Goals'].cumsum().plot(marker='.')
title('Cumulative Goals')
legend(['Suarez','Dzeko','RVP','Aguero','Rooney','Balotelli'],'upper left')
```

```python
df_Suarez['GOPtoSOT'].plot(marker='.')
df_Dzeko['GOPtoSOT'].plot(marker='.')
df_RVP['GOPtoSOT'].plot(marker='.')
title('Goals (Open Play) to SOT Ratio')
legend(['Suarez','Dzeko','RVP'],'upper left')
```

```python
df_Suarez['SOTtoTouF3rd'].plot(marker='.')
df_Dzeko['SOTtoTouF3rd'].plot(marker='.')
df_RVP['SOTtoTouF3rd'].plot(marker='.')
title('SOT to Final Third Touches')
legend(['Suarez','Dzeko','RVP'],'upper left')
```

```python
df_Suarez['SOTtoTou'].plot(marker='.')
df_Dzeko['SOTtoTou'].plot(marker='.')
df_RVP['SOTtoTou'].plot(marker='.')
title('SOT to All Touches')
legend(['Suarez','Dzeko','RVP'],'upper left')
```
