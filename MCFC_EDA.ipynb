# EDA for MCFC dataset

```python
import pandas as pd
```

```python
# Import the Manchester City dataset for 2011-2012 EPL season
df_MCFC = pd.read_excel('mcfc_data/Premier League 2011-12 Match by Match.xls', sheetname='Players')
```

```python
# List all the column names
list(df_MCFC.columns)
```

```python
# Some season totals
df_Player_Season_Totals = df_MCFC.groupby('Player ID').sum()
df_Team_Season_Totals = df_MCFC.groupby('Team').sum()
```

```python
plot(df_Player_Season_Totals['Touches open play opp box'],df_Player_Season_Totals['Goals Open Play'],'.')
```

```python
plot_Team_Tou_G = df_Team_Season_Totals.plot(x='Total Successful Passes All',y='Goals Open Play',marker='.',linestyle='None')
for i, txt in enumerate(df_Team_Season_Totals.index):
    plot_Team_Tou_G.annotate(txt, (df_Team_Season_Totals['Total Successful Passes All'].iloc[i],df_Team_Season_Totals['Goals Open Play'].iloc[i]), size=10, xytext=(0,0), ha='right', textcoords='offset points')
```
