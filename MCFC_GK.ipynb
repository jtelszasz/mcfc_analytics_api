# Goalkeeper Analytics - Manchester City FC Dataset

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
```

```python
# Import the Manchester City dataset for 2011-2012 EPL season
df_2011_2012_season = pd.read_excel('mcfc_data/Premier League 2011-12 Match by Match.xls', sheetname='Players')
```

```python
list(df_2011_2012_season.columns)
```

```python
# Break off just the goalkeepers into a dataframe
df_GKs = df_2011_2012_season[df_2011_2012_season['Position Id'] == 1]
```

```python
# Set multi-index using last name and date
df_GKs.index = pd.MultiIndex.from_arrays([df_GKs['Player Surname'],pd.to_datetime(df_GKs['Date'])])

# Sort by date
df_GKs.sortlevel(level=0,axis=0,inplace=True)
```

```python
# Goalkeeper season totals with additional stats
df_GKs_totals = df_GKs.groupby('Player Surname').sum()

# Reduce set to only GK's who played about a third of the season (12 games) or more (1080 minutes)
df_GKs_totals = df_GKs_totals[df_GKs_totals['Time Played']>=1080]
print 'Total GKs Kept:',len(df_GKs_totals)

# Calculate additional stats

# Save stats
df_GKs_totals['Shots On Conceded P90'] = df_GKs_totals['Shots On Conceded']*90 / df_GKs_totals['Time Played']
df_GKs_totals['Save %'] = df_GKs_totals['Saves Made']/df_GKs_totals['Shots On Conceded']
df_GKs_totals['Saves P90'] = df_GKs_totals['Saves Made']*90/df_GKs_totals['Time Played']
df_GKs_totals['Short Range Save %'] = df_GKs_totals['Saves Made from Inside Box']/df_GKs_totals['Shots On Conceded Inside Box']
df_GKs_totals['Long Range Save %'] = df_GKs_totals['Saves Made from Outside Box']/df_GKs_totals['Shots On Conceded Outside Box']
df_GKs_totals['Big Chances Faced P90'] = df_GKs_totals['Big Chances Faced']*90 / df_GKs_totals['Time Played']

# Distribution stats
df_GKs_totals['Dist P90'] = df_GKs_totals['GK Distribution']*90 / df_GKs_totals['Time Played']
df_GKs_totals['Successful Dist P90'] = df_GKs_totals['GK Successful Distribution']*90/ df_GKs_totals['Time Played']                                    
df_GKs_totals['Successful Dist %'] = df_GKs_totals['GK Successful Distribution'] / df_GKs_totals['GK Distribution']

# Cross stats
# (are "crosses not claimed" actually potential catches for a GK?)
df_GKs_totals['Total Crosses Faced'] = df_GKs_totals['Crosses not Claimed'] + df_GKs_totals['Punches'] + df_GKs_totals['Catches'] + df_GKs_totals['Drops']
df_GKs_totals['Total Crosses Faced P90'] = 90*df_GKs_totals['Total Crosses Faced']/df_GKs_totals['Time Played']
df_GKs_totals['Catch %'] = df_GKs_totals['Catches'] / df_GKs_totals['Total Crosses Faced']
df_GKs_totals['Punch %'] = df_GKs_totals['Punches'] / df_GKs_totals['Total Crosses Faced']
df_GKs_totals['Drop %'] = df_GKs_totals['Drops'] / df_GKs_totals['Total Crosses Faced']
df_GKs_totals['Unclaimed %'] = df_GKs_totals['Crosses not Claimed'] / df_GKs_totals['Total Crosses Faced']

```

```python
# Sort and show table
on = 'Total Crosses Faced P90'
df_GKs_totals.sort(columns=on,inplace=True,ascending=False)

# Save the table to a csv
pd.DataFrame(df_GKs_totals[on]).to_csv('MCFC_GK_saved_output/'+on+'.csv',float_format='%.3f',encoding="iso-8859-15")

df_GKs_totals[[on]]

```
